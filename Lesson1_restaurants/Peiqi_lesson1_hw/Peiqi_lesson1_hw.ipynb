### Q3: Are there any cuisines for which the variance of scores is very high? Why do you think that might be the case?

The whole data file is too huge for my poor computer, so I take the first 1000 row as a sample of the data.

```python
# create array of cuisine code and cuisine name
all_cuisine = mlab.csv2rec("data/Cuisine.txt")
cuisine_code = []
cuisine_name = []
for everyone in all_cuisine:
    x,y = everyone
    cuisine_code.append(x)
    cuisine_name.append(y)
print cuisine_code
print cuisine_name
```

```python
# create numarray for all cuisine
num_cuisines = size(cuisine_code)
scores_of_all_cuisine = zeros((num_cuisines,1000),int) # create 1000 resturants for each type of cuisine
```

```python
# sort restaruants by cuisine_code
all_restaruants = mlab.csv2rec("data/WebExtract_first_1000_rows.txt")
count_4 = 0
for rest in all_restaruants:
    count_5 = 0
    for code in cuisine_code:
        # rest[7] is cuisine code
        if rest[7] == code:
            scores_of_all_cuisine[count_5, count_4] = rest[11] # resr[11] is the score
        count_5 += 1
    count_4 += 1
print scores_of_all_cuisine[15] # print the "Jewish/Kosher" cuisine as a sample
```

```python
# find the std of each cuisine
std_all = []
for each in scores_of_all_cuisine:
    std_all.append(numpy.std(each))
print std_all[15] # print the std of "Jewish/Kosher" cuisine as a sample
```

```python
# draw the std graph of all cuisine
plt.plot(std_all)
```

This the graph of all cuisines's Standard deviation.

```python
# find the cuisine with highest std
max_std = max(std_all)
cuisine_index = std_all.index(max_std)
print cuisine_name[cuisine_index]
```

It turned out that "American" score has the highest Standard deviation, which means the variance of score for "American" cuisine is very high. This may because the munber of American resturants is biggest.

### Q4: Pick another question: What's the most hygenic food?

```python
grade_of_all_cuisine = zeros((num_cuisines,1000),str)
count_7 = 0
for rest in all_restaruants:
    count_8 = 0
    for code in cuisine_code:
        # rest[7] is cuisine code
        if rest[7] == code:
            grade_of_all_cuisine[count_8, count_7] = rest[12] # resr[12] is the current grade
        count_8 += 1
    count_7 += 1
print grade_of_all_cuisine[15] # print the "Jewish/Kosher" cuisine as a sample
```

The NYC resturants letter grade system only have "A", "B", "C" grades. Each "A", "B", "C" earns 3,2,1 points. I assumed "Z" is failed to pass the exam, so it earns -3 points.

```python
# transfer the letter grade to scores: A=3, B=2, C=1
grade_all = []
for each in grade_of_all_cuisine:
    sum_grades = 0
    count_9 = 0
    for each_grade in each:
        if each_grade == 'A':
            sum_grades += 3
            count_9 += 1
        elif each_grade == 'B':
            sum_grades += 2
            count_9 += 1
        elif each_grade == 'C':
            sum_grades += 1
            count_9 += 1
        elif each_grade == 'Z':
            sum_grades -= 3
            count_9 += 1
        else:
            pass
    if count_9 != 0:
        grade_all.append(float(sum_grades)/count_9)
    else:
        grade_all.append(0)
print grade_all[15] # print the numeric grade of "Jewish/Kosher" cuisine as a sample
```

```python
# draw the numeric grade of all types of cuisines
plt.plot(grade_all)
```

```python
# find  the highest grade
print max(grade_all)
```

```python
# find all cuisines with highest grade
max_grade = max(grade_all)
count_10 = 0
for every in grade_all:
    if every == max_grade:
        print cuisine_name[count_10]
    else:
        pass
    count_10 += 1
```
