# My Fitbit Data

```python
fitbit_data = mlab.csv2rec("data/fitbit.csv")
```

```python
fitbit_data[0]
```

```python
dates = []
steps = []
sedentary_minutes = []
light_active_minutes = []
fairly_active_minutes = []
veryactive_minutes = []
minutes_asleep = []

for fit in fitbit_data:
    dates.append(fit[0])
    steps.append(fit[1])
    sedentary_minutes.append(fit[7])
    light_active_minutes.append(fit[8])
    fairly_active_minutes.append(fit[9])
    veryactive_minutes.append(fit[10])
    minutes_asleep.append(fit[12])
```

```python
minutes_accurate = []

for minutes in minutes_asleep:
    if(minutes != 0):
        minutes_accurate.append(minutes)


print numpy.mean(minutes_accurate)/60       
print numpy.median(minutes_accurate)/60
```
