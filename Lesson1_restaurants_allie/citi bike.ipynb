```python
import urllib
import matplotlib
import pylab
```

```python
data_url = "http://citibikenyc.com/stations/json/"

data_handler = urllib.urlopen(data_url)
```

```python
dir(data_handler)
```

```python
citistations = data_handler.read()
```

```python
citistations
```

```python
import json
```

```python
citistation_data = json.loads(citistations)
```

```python
citistation_data
    
```

```python
print len(citistation_data['stationBeanList']) #prints number of stations

num_docks = []
for station in citistation_data['stationBeanList']:
    num_docks.append(station['totalDocks'])
    
    print len(num_docks)
    
```
